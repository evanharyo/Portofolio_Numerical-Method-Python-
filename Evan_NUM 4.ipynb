### Penjelasan Program
Program ini digunakan untuk menyelesaikan masalah juki untuk mendapatkan luas tanah yang pas untuk dirinya

```python
from numpy import cos, exp #mengimpor modul yang di butuhkan untuk menjalankan program
import random
import matplotlib.pyplot as plt
import numpy as np



X = [] #membuat list kosong
Y = []
intX = []
intY = []
juki = []

def bos(n):  #mendefinisikan fungsi bos agar bisa dipanggil di cell berikutnya
    for i in range(n):
        var1 = random.randint(1,9)
        a = str(var1) #membuat nomor acak 
        var2 = random.randint(1,9)
        b = str(var2)#membuat nomor acak 

        formula = "0.5 + exp(-({0}*x**2))*abs(cos({1}*x))".format(a,b)
        def f(x): #mendefinisikan fungsi f
            return eval(formula) #mengeluarkan hasil fungsi
        print('variabel a={0} dan b={1}'.format(a,b))
        print(formula) #mengeluarkan hasil sebelumnya
        
        X = [] 
        Y = []

        for j in np.linspace(0,1,100): #membuat list dengan range 0-1 dibagi sebanyak 100
            X.append(j) #memasukkan titik x ke list x
            Y.append(f(j)) #memasukkan hasil f(x) ke list y

        intX = [] 
        intY = []

        for z in np.linspace(0,1,100): #membuat list dengan range 0-1 dibagi sebanyak 100
            h = z/100
            sum1 = 0
            sum2 = 0
            for j in range(1, int(100/2)):
                sum1 += f(2*j*h)
                sum2 += f((2*j-1)*h)
                sum2 += f(1-h)
            intX.append(z)
            intY.append((h/3)*(f(0)+2*sum1+4*sum2+f(z)))

        value = ((1/100)/3)*(f(0)+2*sum1+4*sum2+f(1))
        print('Luas tanah yang didapat adalah ={0:.5f}'.format(value)) #mengeluarkan hasil luas tanah
        juki.append(value)
        
        plt.xlim(0,1) #membuat plotting dengan range tertentu untuk sumbu x dan y
        plt.ylim(0,2)
        plt.plot(X,Y) #plot titik x y
        plt.plot(intX,intY) #plot titik integral
        plt.legend(['f(x)','integral f(x)'])        
        plt.show()
        print('Luas yang akan didapatkan juki adalah= ',max(juki),'hektar')#mengeluarkan hasil luas juki
        print('')
```
