### Penjelasan Program
Program ini dibuat untuk menyelesaikan SPL dengan menggunakan metode Gauss-Seidel. Ini dijalankan dengan meminta input yaitu matrix yang ingin dicari, solusi awalnya, dan juga toleransi yang ingin digunakan untuk menyelesaikan SPL. Hasil yang diberikan program adalah berapa banyak iterasi SPL dan juga solusi dari SPL itu sendiri.

### Ide Program
Ide nya adalah dengan mengimpor modul yang akan digunakan untuk menjalankan program seperti numpy, lalu meminta input untuk matriks, solusi awal, dan toleransi yang diinginkan pengguna program. Kemudian dilanjutkan dengan menggunakan coding atau program untuk metode Gauss-Seidel yang sudah diberikan pada salah satu praktikum dan juga dengan menambahkan program untuk membuat tabel, sebelum di akhir mengeprint aproksimasi dari solusi SPL serta tabelnya untuk memberi tahu iterasinya.

### Penjelasan Algoritma
- Mengimpor modul numpy dan tabulate agar bisa menjalankan program berikutnya
- Meminta untuk menginput matrix,intial atau nilai awal, dan juga tol yang berupa toleransi yang diinginkan 
- Mendefinisikan program untuk metode Gauss Seidel dengan nama Gauss 
- Memberi tambahan coding untuk membuat tabel
- Mengeprint tabel iterasi
- Mengeprint hasil iterasi
- Mengeluarkan hasil dari pendefinisian metode Gauss Seidel yang sudah di buat sebelumnya untuk mengeluarkan hasil dari SPL yang ditentukan di awal 


## No 1


```python
import numpy as np                                      #mengimpor modul untuk digunakan di program
from tabulate import tabulate                           #mengimpor modul untuk membuat tabel

                                                        #meminta memasukkan input untuk matrix, intial solution, dan toleransi
matr = np.array(eval(input('Masukkan matrix yang anda ingin selesaikan ! : ')))
initial = np.array(eval(input('Masukkan solusi pertama yang ingin ditetapkan: ')),dtype=float)
tol = eval(input('Masukkan batas toleransi yang diinginkan: '))

def Gauss(matrix,initial,tol):                          #mendefinisikan metode gauss seidel dan membuat tabel
  n = np.shape(matrix)[0]                               #membuat ukuran matriks sesuai byk variabel
  x0 = initial                                          #menjadikan x0 
  x = x0.copy()                                         #mendefinisikan x=x0 dan gunakan copy agar jika x0 berubah x nya tidak
  header = [f'X{i+1}' for i in range (n)]               #membuat judul header sesuai dgn x yang dicari
  header.insert(0, 'Iterasi')                           #membuat header dengan judul sesuai x dan di masukkan dari kolom pertama
  table = [header]
  loop1 = []                                            #membuat list kosong untuk diisi hasil iterasi pertama
  iterasi = 1
  loop1.append(iterasi)                                 #memasukkan iterasi ke loop
  for i in range(n):                                    #metode gauss seidel
    b = matrix[i,n]                                     #membuat konstanta pada SPL
    for j in range(n):
      if j!=i:
        if j<i:
          b-=matrix[i,j]*x[j]
        else:
          b-=matrix[i,j]*x0[j]
    x[i] = b/matrix[i,i]                                 #hasil aproksimasi gauss seidel
    loop1.append(x[i])
  table.append(loop1)                                    #memasukkan ke tabel
  iterasi += 1
  
  while np.linalg.norm(x-x0,np.inf)>=tol:                #cek kriteria penyetop menggunakan error mutlak
    x0=x                                                 #definisikan x0 sebagai x(k-1)
    x = x0.copy()                                        #definisikan x sebagai x(k), namun awalnya samakan dengan x0
    loop2 = []                                           #membuat list kosong untuk iterasi berikut berikutnya
    loop2.append(iterasi)
    for i in range(n):                                   #metode gauss seidel
      b = matrix[i,n]
      for j in range(n):
        if j!=i:
          if j<i:
            b-=matrix[i,j]*x[j]
          else:
            b-=matrix[i,j]*x0[j]
      x[i]=b/matrix[i,i]                                 #hasil aproksimasi dengan gauss seidel
      loop2.append(x[i])
    table.append(loop2)
    iterasi += 1
                                                         #print dan return hasil x 
  print(tabulate(table, headers='firstrow', tablefmt='fancy_grid',numalign="center"))
  print('Aproksimasi solusi SPL tersebut dengan metode Gauss-Seidel adalah :')
  return x                                               #nilai yang direturn fungsi ini adalah x(k)

print(Gauss(matr,initial,tol))                           #mengeluarkan hasil akhir
```

#### Penjelasan nomor 1
Dengan menggunakan program diatas, SPL dikatakan konvergen karena memiliki hasil akir, lebih tepatnya pada iterasi ke-8.
